### __`Upcasting` and `Downcasting` in Numpy arrays__

__Upcasting:__ It is the process of converting data types to a higher precision or larger size to avoid data loss during operations. 
- It is performed automatically by NumPy when combining arrays with different data types.
- For example, when adding an integer array to a float array, the integer array will be upcast to float to maintain precision.

__Downcasting:__ It involves converting data types to a lower precision or smaller size, potentially leading to data loss. 
- NumPy generally avoids downcasting automatically to prevent loss of information.
- However, it can be done explicitly using the .astype() method. Downcasting should be done cautiously, as it can truncate or alter the data.

```python
# Array of integers

array_1 = np.array([[1,2,3],[4,5,6]])
print(array_1)
print(array_1.dtype)
```

```python
# Array of floats

array_2 = np.array([[7.1,8.2,9.1], [10.4,11.2, 12.3]])
print(array_2)
print(array_2.dtype)
```

```python
# int is upcasted to float when performing adition operation

array_3 = array_1 + array_2
print(array_3)
print(array_3.dtype)
```

### __Checking Image dimensions in Numpy__

- A grayscale image, when represented as a NumPy array, typically has two dimensions (height and width) or three dimensions with the last dimension having a size of 1.
- Color images, on the other hand, usually have three dimensions with the last dimension having a size of 3 (for RGB) or 4 (for RGBA).
- 

```python
# Displaying sample image from URL

url = 'https://images.freeimages.com/images/large-previews/d08/path-1400935.jpg?fmt=webp&h=350'

response = requests.get(url)
img = Image.open(BytesIO(response.content))
display(img)
```

```python
# Convert the PIL image to Numpy array

image_arr = np.array(img)
print(image_arr.shape)
print(image_arr.dtype)
print(image_arr.ndim)
```
